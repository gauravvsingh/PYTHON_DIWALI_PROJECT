```python
import numpy as np
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as mpl
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.rename(columns={'Marital_Status':'Shaddi'})
```

```python
df.describe()
```

```python
#EXPLORATORY DATA ANALYTICS
#GENDER
```

```python
ax=sea.countplot(x='Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sea.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
#From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
#AGE
```

```python
ax=sea.countplot(data = df ,x = 'Age Group',hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Age Group Vs Amount
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending= False)
sea.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
#From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
#STATES
```

```python
# Sales Vs Orders
sales_state = df.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by = 'Orders',ascending= False).head(10)
sea.set(rc={'figure.figsize':(17,5)})
sea.barplot(data=sales_state,x='State',y='Orders')
```

```python
# States Vs Orders
sales_state = df.groupby(['State'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending= False).head(10)
sea.set(rc={'figure.figsize':(17,5)})
sea.barplot(data=sales_state,x='State',y='Amount')
```

```python
#MARITAL STATUS
```

```python
ax = sea.countplot(data=df,x='Marital_Status')
sea.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending= False)
sea.set(rc={'figure.figsize':(7,5)})
sea.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

```python
#From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
#OCCUPATION
```

```python
#Occupation Vs Orders
ax = sea.countplot(data=df,x='Occupation')
sea.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending= False)
sea.set(rc={'figure.figsize':(20,5)})
sea.barplot(data=sales_state,x='Occupation',y='Amount')
```

```python
#From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation secto
```

```python
#PRODUCT CATEGORY
```

```python
ax = sea.countplot(data=df,x='Product_Category')
sea.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending= False)
sea.set(rc={'figure.figsize':(30,10)})
sea.barplot(data=sales_state,x='Product_Category',y='Amount')
```

```python
#From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```

```python
#PRODUCT ID VS ORDERS
```

```python
sales_state = df.groupby(['Product_ID'],as_index= False)['Orders'].sum().sort_values(by = 'Orders',ascending= False).head(10)
sea.set(rc={'figure.figsize':(20,5)})
sea.barplot(data=sales_state,x='Product_ID',y='Orders')
```
